```
import pickle
import cv2
import os
import matplotlib.pyplot as plt
```

```
all_images_path = "/content/drive/My Drive/DMW/Training_Images/"
```

```
def zeroes(s):
  if s.endswith('_left.jpg'):
    return s.rjust(13,'0')
  return s.rjust(14,'0')

def strp(s):
  temp = s.lstrip('0')
  if temp.startswith('_'):
    temp = '0' + temp
  return temp
```

```
sort_file = os.listdir(all_images_path)
data = {}

sort_file = list(map(zeroes, sort_file))
sort_file.sort()
sort_file = list(map(strp, sort_file))

count = 0

for image in sort_file:
  count += 1
  temp = cv2.imread(all_images_path + image)
  temp = cv2.resize(temp, (256, 256))
  data[image] = temp
  if count % 100 == 0:
    print(count, 'done')

print(len(data))

myfile = open("/content/drive/My Drive/DMW/pickled_data", mode='wb')
pickle.dump(data, myfile)
myfile.close()
```

```
# to load pickled file
myfile = open("/content/drive/My Drive/DMW/pickled_data", mode='rb')
temp = pickle.load(myfile)
myfile.close()

print(len(temp))
temp = cv2.cvtColor(temp[list(temp.keys())[0]], cv2.COLOR_BGR2RGB)
plt.imshow(temp)
```
