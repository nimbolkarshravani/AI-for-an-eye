```
def classifier(image_data):
  image_data = cv2.resize(image_data, (50, 50))
  image_data_temp = cv2.cvtColor(image_data, cv2.COLOR_BGR2GRAY)

  image_data = keras.preprocessing.image.img_to_array(image_data)
  image_data = tf.expand_dims(image_data, 0)

  image_data_temp = keras.preprocessing.image.img_to_array(image_data_temp)
  image_data_temp = tf.expand_dims(image_data_temp, 0)
  
  if float(classifier_L_or_R.predict(image_data_temp)) < 0.5:
    print('Image is Left Fundus')
    if float(classifier_L_N.predict(image_data)) > 0.5:
      print('Normal')
    else:
      if float(classifier_L_A.predict(image_data)) > 0.5:
        print('AMD: positive')
      else:
        print('AMD: negative')
      if float(classifier_L_C.predict(image_data)) > 0.5:
        print('Cataract: positive')
      else:
        print('Cataract: negative')
      if float(classifier_L_G.predict(image_data)) > 0.5:
        print('Glaucoma: positive')
      else:
        print('Glaucoma: negative')
      if float(classifier_L_H.predict(image_data)) > 0.5:
        print('Hypertension: positive')
      else:
        print('Hypertension: negative')
      if float(classifier_L_M.predict(image_data)) > 0.5:
        print('Myopia: positive')
      else:
        print('Myopia: negative')
  else:
    print('Image is Right Fundus')
    if float(classifier_R_N.predict(image_data)) > 0.5:
      print('Normal')
    else:
      if float(classifier_R_A.predict(image_data)) > 0.5:
        print('AMD: positive')
      else:
        print('AMD: negative')
      if float(classifier_R_C.predict(image_data)) > 0.5:
        print('Cataract: positive')
      else:
        print('Cataract: negative')
      if float(classifier_R_G.predict(image_data)) > 0.5:
        print('Glaucoma: positive')
      else:
        print('Glaucoma: negative')
      if float(classifier_R_H.predict(image_data)) > 0.5:
        print('Hypertension: positive')
      else:
        print('Hypertension: negative')
      if float(classifier_R_M.predict(image_data)) > 0.5:
        print('Myopia: positive')
      else:
        print('Myopia: negative')
  print('\n')
```

```
classifier(image_data=img1)
classifier(image_data=img2)
```
